```python
import numpy as np
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
```

```python
# Generate random linearly separable data
np.random.seed(42)
X_train = np.random.uniform(-1, 1, (30, 2))
X_train = np.column_stack((np.ones(len(X_train)), X_train))  # Add bias term
true_weights = np.random.uniform(-1, 1, 3)
y_train = np.sign(np.dot(X_train, true_weights))

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 1], X_train[:, 2], c=y_train, cmap=plt.cm.bwr, marker='o', edgecolors='k')
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Linearly Separable Dataset")
plt.grid(True)
plt.show()
```

```python
def perceptron_learning_algorithm(X, y):
    # Initialize weights
    w = np.zeros(X.shape[1])
    converged = False
    updates = 0
    while not converged:
        converged = True
        for i in range(len(X)):
            if np.sign(np.dot(X[i], w)) != y[i]:
                w += y[i] * X[i]
                converged = False
                updates += 1
    return w, updates
```

```python
# Run PLA
final_weights, num_updates = perceptron_learning_algorithm(X_train, y_train)
```
